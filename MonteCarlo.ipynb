```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#assume outcomes can be symmetrically positive or negative
def generate_outcome(win_rate):
    if random.random() < win_rate:
        return 1
    else:
        return -1
```

```python
#calculate the winning probability
def winProb(results, prt=True):
    count_lose = sum(1 for num in results if num < 0)
    length = len(results)
    count_win = length - count_lose
    win_prob = count_win/length
    mean_return = np.mean(results)
    
    if prt:
        print('(n_lose, n_win): ', (count_lose, count_win))
        print('win probability: ', win_prob)
        print('mean return: ', mean_return)
    return (win_prob, mean_return)
```

```python
#monte carlo simulation
def monte_carlo(n_simulation=100000, win_rate=0.5, proportion=0.01, init_capital=10000000, commission=0.0003,
               periods=15, init_leverage=1, multiplier_limit=2**10, goal_profit=0.3):
    results = []
    for c in range(n_simulation):
        cash = init_capital
        value_short = 0
        value_long = 0
        value = init_capital
        multiplier = 2
        
        #day 0
        value_short = ((init_leverage-1)*init_capital)
        cash += value_short*(1-commission)
        value_long = cash*(1-commission)
        cash = 0
        value = value_long-value_short
        
        #after
        for i in range(1, periods, 1):
            outcome1 = generate_outcome(win_rate)
            value_short += outcome1*proportion*value_short
            outcome2 = generate_outcome(win_rate)
            value_long += outcome2*proportion*value_long
            if init_capital > value_long-value_short:
                value_short += (init_leverage*init_capital)*(multiplier/2)
                cash += (init_leverage*init_capital)*(multiplier/2)*(1-commission)
                value_long += cash*(1-commission)
                cash = 0
                if multiplier < multiplier_limit:
                    multiplier *= 2
            value = value_long-value_short
            if (value/init_capital-1)>goal_profit:
                value *= (1-commission)
                break
        results.append(value/init_capital-1)
    return results
```

```python
r = monte_carlo(goal_profit=0.3)
```

```python
winProb(r)
```

## Create plots

```python
#for win rates and mean returns
x_space = range(1,21)
win_l = []
mean_l = []
for i in x_space:
    win_prob, mean_r = winProb(monte_carlo(n_simulation=1000000, init_leverage=i), prt=False)
    win_l.append(win_prob)
    mean_l.append(mean_r)
```

```python
plt.plot(x_space,np.array(win_l)*100)
plt.xlabel('Times Initial Leverage')
plt.ylabel('Winning Probability %')
plt.xticks(x_space)
plt.grid()

plt.savefig('figures/win_prob.png',bbox_inches='tight')
plt.show()
```

```python
plt.plot(x_space,np.array(mean_l)*100)
plt.xlabel('Times Initial Leverage')
plt.ylabel('Mean Returns %')
plt.xticks(x_space)
plt.grid()

plt.savefig('figures/mean_return.png',bbox_inches='tight')
plt.show()
```

## Simulations for opponents

```python
#assume they maximize their risk by all in long position
#assume they have good luck so that oppo_win_rate is 0.55
n_simulation=10000
c_l=[]
for c in range(n_simulation):
    oppo_win_rate = 0.55
    proportion = 0.02
    init_capital = 10000000
    comission = 0.0003
    periods=15
    value_long = init_capital

    for i in range(1, periods, 1):
        outcome2 = generate_outcome(oppo_win_rate)
        value_long += outcome2*proportion*value_long
    c_l.append(value_long/init_capital-1)
```

```python
print('max profit', np.max(c_l))
```
